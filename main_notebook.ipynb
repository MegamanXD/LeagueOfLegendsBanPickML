```python
synergy_counters_df['best_partners'] = [champion_list.split(",") for champion_list in synergy_counters_df['best_partners']]
synergy_counters_df['worst_counters'] = [champion_list.split(",") for champion_list in synergy_counters_df['worst_counters']]

synergy_counters_df
```

```python
synergy_counters_df['my_champion'].unique()
```

```python
synergy_counters_df['my_champion'] = [champion_name_id_dict.get(champion) for champion in synergy_counters_df['my_champion']]

synergy_counters_df
```

```python
synergy_counters_df.isna().sum()
```

```python
converted_list = []

for champion_list in synergy_counters_df['best_partners']:
    row_list = []
    for champion in champion_list:
        row_list.append( champion_name_id_dict.get(champion) )
    converted_list.append(row_list)

synergy_counters_df['best_partners'] = converted_list
synergy_counters_df
```

```python
converted_list = []

for champion_list in synergy_counters_df['worst_counters']:
    row_list = []
    for champion in champion_list:
        row_list.append( champion_name_id_dict.get(champion) )
    converted_list.append(row_list)

synergy_counters_df['worst_counters'] = converted_list
synergy_counters_df
```
