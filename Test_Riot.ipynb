```python
def scrape_one_match(player_puuid, region_new, match_id):
    match_detail = watcher.match.by_id(region_new, match_id)
    match_participants = match_detail['info']['participants']
#     player_win_loss = False

    for row in match_participants:
        if row['puuid'] == player_puuid:
            player_champ_id = row['championId']
            player_champ_name = row['championName']
            player_role = row['teamPosition']
            player_team = row['teamId']
            player_win_loss = row['win']
            
    ally_champs = []
    ally_champ_names = []
    ally_champ_position = []
    ally_win_loss = player_win_loss

    enemy_champs = []
    enemy_champ_names = []
    enemy_champ_position = []
    enemy_win_loss = not player_win_loss

    for row in match_participants:
        if row['teamId'] == player_team:
            ally_champs.append( row['championId'] )
            ally_champ_names.append( row['championName'] )
            ally_champ_position.append( row['teamPosition'] )

        if row['teamId'] != player_team:
            enemy_champs.append( row['championId'] )
            enemy_champ_names.append( row['championName'] )
            enemy_champ_position.append( row['teamPosition'] )
    
    ally_dictionary = {}
    enemy_dictionary = {}
    champion_dictionary = {}
    win_loss_dictionary = {}
    position_dictionary = {}

    for i in range(5):
        ally_dictionary.update({ally_champ_names[i] : ally_champs[i]})
        win_loss_dictionary.update({ally_champ_names[i] : ally_win_loss})
        position_dictionary.update({ally_champ_names[i] : ally_champ_position[i]})

        enemy_dictionary.update({enemy_champ_names[i] : enemy_champs[i]})
        win_loss_dictionary.update({enemy_champ_names[i] : enemy_win_loss})
        position_dictionary.update({enemy_champ_names[i] : enemy_champ_position[i]})    
    champion_dictionary = ally_dictionary.copy()
    champion_dictionary.update(enemy_dictionary)
        
    from itertools import combinations
    ally_combinations = list(combinations(ally_dictionary, 3))
    enemy_pairs = list(combinations(enemy_dictionary, 2))
    
    all_ally_combinations = [list(an_ally_combination) for an_ally_combination in ally_combinations]
    all_enemy_combinations = [list(an_enemy_pair) for an_enemy_pair in enemy_pairs]
    
    data_list = []

    player_champ_id = None
    player_role = None
    player_position = None
    ally_champ1 = None
    ally_champ1_name = None
    ally_champ1_position = None
    ally_champ2 = None
    ally_champ2_name = None
    ally_champ2_position = None
    enemy_champ1 = None
    enemy_champ1_name = None
    enemy_champ1_position = None
    enemy_champ2 = None
    enemy_champ2_name = None
    enemy_champ2_position = None

    for an_ally_combination in all_ally_combinations:
        for an_enemy_pair in all_enemy_combinations:
            player_champ = an_ally_combination[0]
            player_champ_id = champion_dictionary.get(player_champ)
            player_position = position_dictionary.get(player_champ)
            play_win_loss = win_loss_dictionary.get(player_champ)
            ally_champ1 = an_ally_combination[1]
            ally_champ1_id = champion_dictionary.get(ally_champ1)
            ally_champ1_position = position_dictionary.get(ally_champ1)
            ally_champ2 = an_ally_combination[2]
            ally_champ2_id = champion_dictionary.get(ally_champ2)
            ally_champ2_position = position_dictionary.get(ally_champ2)
            enemy_champ1 = an_enemy_pair[0]
            enemy_champ1_id = champion_dictionary.get(enemy_champ1)
            enemy_champ1_position = position_dictionary.get(enemy_champ1)
            enemy_champ2 = an_enemy_pair[1]
            enemy_champ2_id = champion_dictionary.get(enemy_champ2)
            enemy_champ2_position = position_dictionary.get(enemy_champ2)

            data = [play_win_loss, player_champ, player_champ_id, player_position, ally_champ1, ally_champ1_id, ally_champ1_position, ally_champ2, ally_champ2_id, ally_champ2_position, enemy_champ1, enemy_champ1_id, enemy_champ1_position, enemy_champ2, enemy_champ2_id, enemy_champ2_position]
            data_list.append(data)

    data_list = pd.DataFrame(data_list, columns = ['is_player_win', 'player_champ', 'player_champ_id', 'player_position', 'ally_champ1', 'ally_champ1_id', 'ally_champ1_position', 'ally_champ2', 'ally_champ2_id', 'ally_champ2_position', 'enemy_champ1', 'enemy_champ1_id', 'enemy_champ1_position', 'enemy_champ2', 'enemy_champ2_id', 'enemy_champ2_position'])
    return data_list
```

```python
def scrape_all_matches(api_key, player_name, region_new, region_old):
    watcher = LolWatcher(api_key)
    player_found = watcher.summoner.by_name(region_old, player_name)
    my_ranked_stats = watcher.league.by_summoner(region_old, player_found['id'])
    player_puuid = player_found.get("puuid")

    player_matches = watcher.match.matchlist_by_puuid(region_new, player_puuid)
    train_data = pd.DataFrame(columns = ['is_player_win', 'player_champ', 'player_champ_id', 'player_position', 'ally_champ1', 'ally_champ1_id', 'ally_champ1_position', 'ally_champ2', 'ally_champ2_id', 'ally_champ2_position', 'enemy_champ1', 'enemy_champ1_id', 'enemy_champ1_position', 'enemy_champ2', 'enemy_champ2_id', 'enemy_champ2_position'])
    
    for match_id in player_matches:
        data = scrape_one_match(player_puuid, region_new, match_id)
        train_data = pd.concat([train_data, data], axis=0)

        
    return train_data
```
